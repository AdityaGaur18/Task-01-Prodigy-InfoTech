**Importing all the libraries**

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

**Reading the Data**


```python

df = pd.read_csv("population.csv")
```

```python
df
```

```python
df.describe
```

```python
# Extraction of relevant column

data_2020 = df.loc[:, ['Country Name', '2022']]
```

**Removing the missing/null value**

```python
data_2020 = data_2020.dropna().head()
```

**Use Histogram to Visualize the Data**

```python
sns.set(style="whitegrid")

# Histogram for population distribution in 2020

plt.figure(figsize = (12, 6))
sns.histplot(data = data_2020, x='2022', bins=30, kde=True, color='red')
plt.title('Population Distribution in 2022')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

**Visualizing using Bar Chart**

---



```python


# Convert population column to numeric (in case it isn't already)
data_2020['2022'] = pd.to_numeric(data_2020['2022'], errors='coerce')

# Sort data by population in descending order and select top 10
top_10_countries = data_2020.sort_values(by='2022', ascending=False).head(10)

# Set up the visual style
sns.set(style="whitegrid")

# Bar chart for top 10 most populous countries in 2022
plt.figure(figsize=(12, 6))
sns.barplot(data = top_10_countries, x='Country Name', y='2022', palette='pastel')
plt.title('Top 10 Most Populous Countries in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
